Changed the id coln to float 

```python
# pre_df["id"].apply(lambda x: int(x) if isinstance(x,str) and x.isnumeric() else None )
pre_df["id"] = pd.to_numeric(pre_df["id"], errors="coerce")

pre_df.info()
```

```python
pre_df["deviceId"] = pd.to_numeric(pre_df["deviceId"], errors="coerce")
pre_df["distance"] = pd.to_numeric(pre_df["distance"], errors="coerce")
pre_df["totalDistance"] = pd.to_numeric(pre_df["totalDistance"], errors="coerce")

pre_df.info()
```

```python
cols = [
    "alarm", "ignition", "status", "motion", "deviceId",
    "type", "protocol", "outdated", "valid", "altitude",
    "address", "accuracy", "network"
]

for col in cols:
    print(f"\n{col} has these unique values:{pre_df[col].nunique()}")
    print(pre_df[col].unique())
```

```python
pre_df["ignition"] = pre_df["ignition"].apply(lambda x: False if x == "FALSE" else x)
pre_df["ignition"].unique()
```

```python
pre_df["status"] = pre_df["status"].apply(lambda x : int(x) if isinstance(x , str) and x.isnumeric() else x)
```

```python
pre_df["status"].unique()
```

Converts:

True, "TRUE" → True

False, "FALSE" → False

column-name junk → NaN

```python
bool_cols = ["ignition", "motion", "outdated", "valid"]

for col in bool_cols:
    pre_df[col] = (
        pre_df[col]
        .astype(str)
        .str.lower()
        .map({"true": True, "false": False})
    )
```

✔ Converts:

"4294949887" → 4294949887

"140" → 140

"status", "accuracy" → NaN

```python
num_cols = ["status", "deviceId", "altitude", "accuracy"]

for col in num_cols:
    pre_df[col] = pd.to_numeric(pre_df[col], errors="coerce")
```


✔ Removes values like:

"alarm" from alarm

"protocol" from protocol

```python
str_cols = ["alarm", "type", "protocol", "address", "network"]

for col in str_cols:
    pre_df[col] = pre_df[col].replace(col, pd.NA)
```
